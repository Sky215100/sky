```python
df = pd.read_csv('C:/Users/USER/Desktop/python/新增資料夾/disney_movies_total_gross.csv', parse_dates=['release_date'])
df.head()
```

```python
gross_adj = df.sort_values(by = 'inflation_adjusted_gross', ascending = False)
gross_adj.head()
```

```python
#plot which movies net highest profit
gross_adj10 = gross_adj.head(10)
plt.figure(figsize=(10,10))
sns.barplot(x=gross_adj10.movie_title,
            y=gross_adj10.inflation_adjusted_gross, 
           palette="Paired")
plt.xlabel("Movie", size = 15)
plt.ylabel("Gross Income (billions)", size = 15)
plt.title("Disney Movie Incomes Adjusted for Inflation", size = 18)
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

```python
df['year'] = pd.DatetimeIndex(df['release_date']).year
genrey = df.groupby(['genre', 'year']).mean()
genre2 = genrey.reset_index()
genre2['inflation_adjusted_gross'] = genre2['inflation_adjusted_gross'].astype(int)
genre2 = genre2.sort_values(by = 'inflation_adjusted_gross', ascending = False)
genre2.head(15)
```

```python
#Plot Time Series for gross income over time

plt.figure(figsize=(10,10))
sns.lineplot(x=genre2.year,
            y=genre2.inflation_adjusted_gross)
plt.xlabel("Year", size = 15)
plt.ylabel("Gross Income (billions)", size = 15)
plt.title("Disney Movie Profits Over Time Adjusted for Inflation", size = 20)
plt.xticks(rotation = 80)
plt.tight_layout()
plt.show()
```
